```python
import pandas as pd

# Load the data
file_path = '/content/updated_merged_data_19102024.xlsx'  # Replace with your file path
df = pd.read_excel(file_path)

# Group by 'userid' and 'attainment' and apply a summary (e.g., sum, count, etc.)
grouped_df = df.groupby(['userid', 'attainment']).sum()  # Use sum() or another function depending on your needs

# Display or save the grouped data
print(grouped_df)
# If you'd like to save the grouped data to a new Excel file
grouped_df.to_excel('grouped_data.xlsx')
```
