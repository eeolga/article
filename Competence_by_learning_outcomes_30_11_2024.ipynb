```python
import pandas as pd

# Input data
subjects = [
    {"Program": "Applied Mathematics", "ECTS": 6, "Purpose": "Fundamental to nearly all areas of natural and engineering sciences, logical and mathematical thinking ability",
     "Competencies": {
         "1-Understands": ["Arithmetic operations with matrices, determinants up to the fifth order", "Representations, classes, types of functions of one real variable, domain and range"],
         "2-Evaluates": ["Use matrix algebra methods to solve systems of linear equations", "Manipulate vectors in two and three dimensions, solve geometrical problems", "Find derivatives of elementary functions"],
         "3-Implements": ["Basic techniques of integration"]
     }},
    {"Program": "Engineering Graphics", "ECTS": 3, "Purpose": "Knowledge of current standards for making and reading drawings, spatial imagination ability to prepare and read drawings",
     "Competencies": {
         "1-Understands": ["Standards and reference material handling", "Norms concerning formatting and techniques of representation"],
         "2-Evaluates": ["Read and supplement drawings", "Acquire skills of drawing techniques"],
         "3-Implements": ["Prepare drawings in accordance with applicable norms"]
     }},
    {"Program": "Basics of CAD Systems", "ECTS": 3, "Purpose": "General principles of the computer program AutoCAD 2D drawing and the basic drawing techniques required",
     "Competencies": {
         "1-Understands": ["Necessary AutoCAD drawing techniques"],
         "2-Evaluates": ["Prepare drawings and form them according to ISO standards"],
         "3-Implements": ["Process and convert drawings for transmission and presentation"]
     }},
    {"Program": "Applied Physics", "ECTS": 6, "Purpose": "Knowledge of the general laws of physics in the fields of mechanics, molecular physics, thermodynamics, and electricity",
     "Competencies": {
         "1-Understands": ["Physics basic concepts and relationships"],
         "2-Evaluates": ["Explain physical constitutions and processes, connect to modern technology"],
         "3-Implements": ["Solve physics problems with correct formulas and unit conversions", "Perform practical work with relevant measurements and calculations", "Prepare laboratory work reports"]
     }}
]

# Calculate ECTS distribution for competencies
results = []

for subject in subjects:
    total_ects = subject['ECTS']
    hours_per_ects = 26
    total_hours = total_ects * hours_per_ects
    competencies = subject['Competencies']

    total_competencies = sum(len(v) for v in competencies.values())
    hours_per_competency = total_hours / total_competencies

    for category, items in competencies.items():
        for item in items:
            results.append({
                "Program": subject['Program'],
                "ECTS": total_ects,
                "Competence Type": category,
                "Competence Definition": item,
                "Hours Allocated": hours_per_competency
            })

# Convert results to a DataFrame
results_df = pd.DataFrame(results)

# Save the DataFrame to an Excel file
# Changed 'df' to 'results_df' to refer to the correct DataFrame
excel_file = 'competencies.xlsx' #Added this line to define excel_file
results_df.to_excel(excel_file, index=False, engine='openpyxl')  # 'openpyxl' is the default engine for XLSX

print(f"File converted to {excel_file}")
```

```python
import pandas as pd
import os

# Input data
subjects = [
    {"Program": "Applied Mathematics", "ECTS": 6,
     "Competencies": [("Understands", 40), ("Evaluates", 30), ("Implements", 30)]},
    {"Program": "Engineering Graphics", "ECTS": 3,
     "Competencies": [("Understands", 40), ("Evaluates", 30), ("Implements", 30)]},
    {"Program": "Basics of CAD Systems", "ECTS": 3,
     "Competencies": [("Understands", 40), ("Evaluates", 30), ("Implements", 30)]},
    {"Program": "Applied Physics", "ECTS": 6,
     "Competencies": [("Understands", 40), ("Evaluates", 30), ("Implements", 30)]}
]

# Function to calculate hours and ECTS for each competency
def calculate_competency_distribution(subjects):
    data = []
    for subject in subjects:
        total_hours = subject["ECTS"] * 26
        for competency, percentage in subject["Competencies"]:
            hours = total_hours * (percentage / 100)
            ects_portion = hours / 26
            data.append({
                "Program": subject["Program"],
                "Competency": competency,
                "Percentage": percentage,
                "ECTS Portion": round(ects_portion, 2),
                "Hours": round(hours, 2)
            })
    return data

# Calculate distribution
distribution = calculate_competency_distribution(subjects)

# Convert to DataFrame
df = pd.DataFrame(distribution)

output_file = "learning_subject_ects_distribution.xlsx"

# Get the current working directory
current_directory = os.getcwd()

# Prepend current directory to the filename to create the full path
output_file_path = os.path.join(current_directory, output_file)

# Create the directory if it doesn't exist (using the full path)
os.makedirs(os.path.dirname(output_file_path), exist_ok=True)

df.to_excel(output_file_path, index=False)  # Save to the full path

print(f"Table saved to {output_file_path}")
```

```python
import pandas as pd

# Input data: Program details and percentages
program_data = [
    ["Applied Mathematics", 6, 156, [
        [1, "Understands", "Arithmetic operations with matrices, determinants up to the fifth order"],
        [1, "Understands", "Representations, classes, types of functions of one real variable, domain and range"],
        [2, "Evaluates", "Use matrix algebra methods to solve systems of linear equations"],
        [2, "Evaluates", "Manipulate vectors in two and three dimensions, solve geometrical problems"],
        [2, "Evaluates", "Find derivatives of elementary functions"],
        [3, "Implements", "Basic techniques of integration"]
    ]],
    ["Engineering Graphics", 3, 78, [
        [1, "Understands", "Standards and reference material handling"],
        [1, "Understands", "Norms concerning formatting and techniques of representation"],
        [2, "Evaluates", "Read and supplement drawings"],
        [2, "Evaluates", "Acquire skills of drawing techniques"],
        [3, "Implements", "Prepare drawings in accordance with applicable norms"]
    ]],
    ["Basics of CAD Systems", 3, 78, [
        [1, "Understands", "Necessary AutoCAD drawing techniques"],
        [2, "Evaluates", "Prepare drawings and form them according to ISO standards"],
        [3, "Implements", "Process and convert drawings for transmission and presentation"]
    ]],
    ["Applied Physics", 6, 156, [
        [1, "Understands", "Physics basic concepts and relationships"],
        [2, "Evaluates", "Explain physical constitutions and processes, connect to modern technology"],
        [3, "Implements", "Solve physics problems with correct formulas and unit conversions"],
        [3, "Implements", "Perform practical work with relevant measurements and calculations"],
        [3, "Implements", "Prepare laboratory work reports"]
    ]]
]

# Percentages for each competence type
competence_percentages = {
    1: 40,  # Understands
    2: 30,  # Evaluates
    3: 30   # Implements
}

# Convert to proportions
competence_proportions = {k: v / 100 for k, v in competence_percentages.items()}

# Prepare output data
output_data = []
for program, ects, hours, competencies in program_data:
    # Compute total hours for each competence type
    type_hours = {k: hours * v for k, v in competence_proportions.items()}
    for comp_type, comp_desc, comp_detail in competencies:
        allocated_hours = type_hours[comp_type]
        output_data.append([program, ects, hours, comp_type, comp_desc, comp_detail, allocated_hours])

# Create a DataFrame and save to a table file
df = pd.DataFrame(output_data, columns=[
    "Program", "ECTS", "Total Hours", "Competence Type", "Competence Description", "Competence Details", "Allocated Hours"
])

# Save to CSV
output_file = "competence_allocation.csv"
df.to_csv(output_file, index=False)

print(f"Data saved to {output_file}")
```
