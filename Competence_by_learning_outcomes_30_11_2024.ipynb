```python
import pandas as pd

# Define the input data as a dictionary
data = {
    "Program": ["Applied Mathematics", "Engineering Graphics", "Basics of CAD Systems", "Applied Physics"],
    "Purpose": [
        "Fundamental to nearly all areas of natural and engineering sciences, logical and mathematical thinking ability",
        "Knowledge of current standards for making and reading drawings, spatial imagination ability to prepare and read drawings",
        "General principles of the computer program AutoCAD 2D drawing and the basic drawing techniques required",
        "Knowledge of the general laws of physics in the fields of mechanics, molecular physics, thermodynamics and electricity",
    ],
    "LO_1": [
        "Ability to perform arithmetic operations with matrices, calculate determinants up to the fifth order.",
        "Overview of standards and knowing how to handle reference material. Knowing the norms concerning the formatting of drawings and techniques of representation",
        "Ability to use the necessary AutoCAD drawing techniques",
        "Defining physics basic concepts and describing, how physics quantities are related to each other",
    ],
    "LO_2": [
        "Ability to use matrix algebra methods in solving systems of linear equations.",
        "Reading and supplementing drawings. Acquires skills of drawing techniques",
        "Knowing how to prepare a drawing, and form it according to the rules of the ISO standard",
        "Explaining physical constitutions and processes in own words, graphically or using real life examples and connecting them to applications in modern technology",
    ],
    "LO_3": [
        "Ability to manipulate vectors in two and three dimensions, and to solve geometrical problems using vectors",
        "Preparing drawings in accordance with the applicable norms",
        "Ability to process and convert drawings for transmission and presentation",
        "Solving physics problems, choosing right formulas and converting physics units",
    ],
    "LO_4": [
        "Familiarity with various representations, classes and types of functions of one real variable, ability to determine their domain and range",
        None,
        None,
        "Performing practical work, solving a given problem with relevant measurements and calculations",
    ],
    "LO_5": [
        "Knowing the basic rules for differentiation and be able to find derivatives of elementary functions",
        None,
        None,
        "Preparing laboratory work reports with its describing, necessary graphics, analyzing and evaluating obtained results",
    ],
    "LO_6": [
        "Familiarity with the basic techniques of integration",
        None,
        None,
        None,
    ],
}

# Create a DataFrame
df = pd.DataFrame(data)

# Function to extract competencies
def extract_competencies(df):
    competencies = {"understands": [], "can_be_able": [], "implements": []}

    for index, row in df.iterrows():
        for col in df.columns[2:]:  # Loop through LO columns
            if pd.notnull(row[col]):
                lo_text = row[col]
                if any(keyword in lo_text.lower() for keyword in ["familiarity", "knowledge of", "overview"]):
                    competencies["understands"].append((row["Program"], lo_text))
                elif any(keyword in lo_text.lower() for keyword in ["ability to", "be able to", "skills of"]):
                    competencies["can_be_able"].append((row["Program"], lo_text))
                elif any(keyword in lo_text.lower() for keyword in ["preparing", "performing", "implementing"]):
                    competencies["implements"].append((row["Program"], lo_text))

    return competencies

# Extract the competencies
competencies = extract_competencies(df)

# Display the results
for category, items in competencies.items():
    print(f"\n{category.upper()}:")
    for program, text in items:
        print(f"- {program}: {text}")
```

```python
import pandas as pd

# Input data
subjects = [
    {"Program": "Applied Mathematics", "ECTS": 6, "Purpose": "Fundamental to nearly all areas of natural and engineering sciences, logical and mathematical thinking ability",
     "Competencies": {
         "1-Understands": ["Arithmetic operations with matrices, determinants up to the fifth order", "Representations, classes, types of functions of one real variable, domain and range"],
         "2-Can/Be Able": ["Use matrix algebra methods to solve systems of linear equations", "Manipulate vectors in two and three dimensions, solve geometrical problems", "Find derivatives of elementary functions"],
         "3-Implements": ["Basic techniques of integration"]
     }},
    {"Program": "Engineering Graphics", "ECTS": 3, "Purpose": "Knowledge of current standards for making and reading drawings, spatial imagination ability to prepare and read drawings",
     "Competencies": {
         "1-Understands": ["Standards and reference material handling", "Norms concerning formatting and techniques of representation"],
         "2-Can/Be Able": ["Read and supplement drawings", "Acquire skills of drawing techniques"],
         "3-Implements": ["Prepare drawings in accordance with applicable norms"]
     }},
    {"Program": "Basics of CAD Systems", "ECTS": 3, "Purpose": "General principles of the computer program AutoCAD 2D drawing and the basic drawing techniques required",
     "Competencies": {
         "1-Understands": ["Necessary AutoCAD drawing techniques"],
         "2-Can/Be Able": ["Prepare drawings and form them according to ISO standards"],
         "3-Implements": ["Process and convert drawings for transmission and presentation"]
     }},
    {"Program": "Applied Physics", "ECTS": 6, "Purpose": "Knowledge of the general laws of physics in the fields of mechanics, molecular physics, thermodynamics, and electricity",
     "Competencies": {
         "1-Understands": ["Physics basic concepts and relationships"],
         "2-Can/Be Able": ["Explain physical constitutions and processes, connect to modern technology"],
         "3-Implements": ["Solve physics problems with correct formulas and unit conversions", "Perform practical work with relevant measurements and calculations", "Prepare laboratory work reports"]
     }}
]

# Calculate ECTS distribution for competencies
results = []

for subject in subjects:
    total_ects = subject['ECTS']
    hours_per_ects = 26
    total_hours = total_ects * hours_per_ects
    competencies = subject['Competencies']

    total_competencies = sum(len(v) for v in competencies.values())
    hours_per_competency = total_hours / total_competencies

    for category, items in competencies.items():
        for item in items:
            results.append({
                "Program": subject['Program'],
                "ECTS": total_ects,
                "Competence Type": category,
                "Competence Description": item,
                "Hours Allocated": hours_per_competency
            })

# Convert results to a DataFrame
results_df = pd.DataFrame(results)

# Save to a CSV file
output_file = "competencies_allocation.csv"
results_df.to_csv(output_file, index=False)

print(f"Results saved to {output_file}")
```
