Subjects "Basics of Ecology" outputs and "Green Transition Action Fields" program outputs correlation

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Define learning outcomes for both subjects
ecology_outcomes = [
    "Knowing general aims of ecology and its main branches",
    "Knowing main theories and central principles of ecology",
    "Knowing the principles of flux of energy and matter in ecosystems",
    "Ability to interpret professional information from viewpoint of ecology",
    "Experience in reading and interpreting ecological literature",
    "Experience in planning research activities"
]

green_outcomes = [
    "Competence in sustainable practices and renewable technologies",
    "Emphasis on energy saving and optimisation of resource use, the ability to cope with the consequences of climate change",
    "Organisation and optimisation of processes for effective waste management"
]

# Compute TF-IDF vectors
vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform(ecology_outcomes + green_outcomes)

# Compute similarity matrix
similarity_matrix = cosine_similarity(vectors[:len(ecology_outcomes)], vectors[len(ecology_outcomes):])

# Find best matches
def match_outcomes(similarity_matrix, ecology_outcomes, green_outcomes):
    for i, row in enumerate(similarity_matrix):
        best_match_idx = np.argmax(row)
        best_match_score = row[best_match_idx]
        print(f"'{ecology_outcomes[i]}' matches best with '\n  -> {green_outcomes[best_match_idx]}' (Similarity: {best_match_score:.2f})")

match_outcomes(similarity_matrix, ecology_outcomes, green_outcomes)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Define the learning outcomes
engineering_graphics = [
    "Overview of standards and knowing how to handle reference material",
    "Knowing the norms concerning the formatting of drawings and techniques of representation",
    "Acquires skills of drawing techniques",
    "Preparing drawings in accordance with the applicable norms",
    "Reading and supplementing drawings"
]

green_transition = [
    "Competence in sustainable practices and renewable technologies",
    "Emphasis on energy saving and optimisation of resource use, the ability to cope with the consequences of climate change",
    "Organisation and optimisation of processes for effective waste management"
]

# Combine all learning outcomes for vectorization
all_outcomes = engineering_graphics + green_transition

# Compute TF-IDF and cosine similarity
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(all_outcomes)
similarity_matrix = cosine_similarity(tfidf_matrix)

# Convert similarity matrix to DataFrame for readability
df_similarity = pd.DataFrame(similarity_matrix,
                             index=["EG1", "EG2", "EG3", "EG4", "EG5", "GT1", "GT2", "GT3"],
                             columns=["EG1", "EG2", "EG3", "EG4", "EG5", "GT1", "GT2", "GT3"])

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df_similarity, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Learning Outcome Similarity Between Subjects")
plt.show()
```

Subjects "Basics of CAD Systems" outputs and "Green Transition Action Fields" program outputs correlation

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Define learning outcomes
cad_outcomes = [
    "Ability to use the necessary AutoCAD drawing techniques",
    "Knowing how to prepare a drawing, and form it according to the rules of the ISO standard",
    "Ability to process and convert drawings for transmission and presentation"
]

green_outcomes = [
    "Competence in sustainable practices and renewable technologies",
    "Emphasis on energy saving and optimisation of resource use, the ability to cope with the consequences of climate change",
    "Organisation and optimisation of processes for effective waste management"
]

# Combine all outcomes
all_outcomes = cad_outcomes + green_outcomes

# Convert text data into TF-IDF features
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(all_outcomes)

# Compute cosine similarity
similarity_matrix = cosine_similarity(tfidf_matrix[:3], tfidf_matrix[3:])

# Visualization
fig, ax = plt.subplots(figsize=(8, 6))
cax = ax.matshow(similarity_matrix, cmap='coolwarm', vmin=0, vmax=1)
plt.colorbar(cax)
ax.set_xticks(range(len(green_outcomes)))
ax.set_yticks(range(len(cad_outcomes)))
ax.set_xticklabels([f'Green LO {i+1}' for i in range(len(green_outcomes))], rotation=45, ha="left")
ax.set_yticklabels([f'CAD LO {i+1}' for i in range(len(cad_outcomes))])
plt.title("Matching Learning Outcomes between CAD and Green Transition Fields")
plt.show()

# Print similarity scores
for i, cad_lo in enumerate(cad_outcomes):
    for j, green_lo in enumerate(green_outcomes):
        print(f"Similarity between '{cad_lo}' and '{green_lo}': {similarity_matrix[i, j]:.2f}")
```
